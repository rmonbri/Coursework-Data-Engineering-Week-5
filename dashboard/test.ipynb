```python
df_types = pd.merge(plant, plant_type, on='plant_type_id', how='outer')
df_type_botanist = pd.merge(df_types, botanist, on='botanist_id', how='outer')
plant_information = pd.merge(
    df_type_botanist, origin, on='location_id', how='outer').sort_values("plant_id")
plant_information
```

```python
def get_measurements(conn) -> pd.DataFrame:
    '''Returns all measurements from the measurement table'''
    query = "SELECT * FROM measurement;"
    measurement = pd.read_sql(query, conn)
    return measurement
measurement = get_measurements(conn)
measurement.dtypes
```

```python
measurement = measurement[measurement['measurement_time'] > datetime.now() - timedelta(hours=24)]
measurement
```

```python
merged_df = pd.merge(measurement, plant_information,
                                  on='plant_id', how='outer')
```

```python
merged_df.sort_values("plant_id")
```

```python
g_df = plant_information.groupby("botanist_name").agg(
    {"plant_id": "count"}).reset_index()
g_df.rename(columns={"botanist_name": "Botanist", "plant_id": "Number of plants"}, inplace=True)
print(g_df)


pie = alt.Chart(g_df).mark_arc(innerRadius=75).encode(
    theta=alt.Theta(field="Number of plants", type="quantitative", stack=True, scale=alt.Scale(
        type="linear", rangeMax=1.5708, rangeMin=-1.5708)),
    color=alt.Color(field="Botanist", type="nominal",
                    scale=alt.Scale(scheme="greens"))
)

# Text inside the arc segments
text = pie.mark_text(radius=110, fontSize=25, fill="black", baseline="middle", align="center").encode(
    text=alt.Text("Number of plants:Q")
)

# Combine pie chart and text
chart = (pie + text).properties(
    title="Plant Distribution by Botanist"
)
chart
```

```python
import altair as alt

pie = alt.Chart(g_df).mark_arc(innerRadius=75).encode(
    theta=alt.Theta(field="Number of plants", type="quantitative", stack=True, scale=alt.Scale(
        type="linear", rangeMax=1.5708, rangeMin=-1.5708)),
    color=alt.Color(
        field="Botanist", type="nominal",
        scale=alt.Scale(scheme="greens"),
        legend=alt.Legend(orient="right", title="Botanist",
                          offset=0)  # Move legend closer
    )
).properties(
    title="Plant Distribution by Botanist"
)

# Text inside the arc segments
text = pie.mark_text(radius=120, fontSize=16, fill="black", baseline="middle", align="center").encode(
    text=alt.Text("Number of plants:Q")
)

# Combine pie chart and text, and adjust the chart size to bring legend closer
chart = (pie + text).properties(
    title="Plant Distribution by Botanist",
    width=350,  # Decrease width to bring legend closer
    height=350
)

chart
```

```python
single_plant_measurement = merged_df[merged_df['plant_id'] == 1].sort_values("measurement_time").reset_index()
single_plant_measurement
```

```python
single_plant_measurement = single_plant_measurement.sort_values('temperature').reset_index()
```

```python
alt.Chart(single_plant_measurement).mark_line().encode(
    x='measurement_time:T',
    y='temperature:Q'
).properties(
    width=2000
)
```

```python
alt.Chart(single_plant_measurement).mark_line().encode(
    x='measurement_time:T',
    y='moisture:Q'
).properties(
    width=2000
)
```

```python
plant_name = single_plant_measurement['plant_name'].iloc[0]
base = alt.Chart(single_plant_measurement).encode(
    alt.X('measurement_time:T').title(None)
).properties(width=800)

temperature_line = base.mark_line(stroke='#e83d17', interpolate='monotone').encode(
    alt.Y('temperature').axis(
        title='Avg. Temperature (°C)', titleColor='#e83d17')
        )

moisture_line = base.mark_line(stroke='#5276A7', interpolate='monotone').encode(
    alt.Y('moisture').axis(
        title='Moisture Content (%)', titleColor='#5276A7')
)

alt.layer(temperature_line, moisture_line).resolve_scale(
    y='independent'
).properties(title=f"Temperature and moisture content of the {plant_name} over the last 24 hours")
```
